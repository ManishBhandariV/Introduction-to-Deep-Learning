```
from prepare_data import parse_seq
import pickle
import tensorflow as tf
%tensorflow_version 2.x

# this is just a datasets of "bytes" (not understandable)
data = tf.data.TFRecordDataset("data.tfrecords")

# this maps a parser function that properly interprets the bytes over the dataset
# (with fixed sequence length 200)
# if you change the sequence length in preprocessing you also need to change it here
data = data.map(lambda x: parse_seq(x, 200))

# a map from characters to indices
vocab = pickle.load(open("data_vocab", mode="rb"))
vocab_size = len(vocab)
# inverse mapping: indices to characters
ind_to_ch = {ind: ch for (ch, ind) in vocab.items()}

print(vocab)
print(vocab_size)
```

```
 #to see what is there in my data
for raw_record in data.take(10):        
  print(repr((raw_record)))               
```

```
batch_size = 128
num_epochs = 20

#dataset = data.shuffle(1000+3*batch_size)
dataset = data.padded_batch(batch_size,drop_remainder=True,
                               padded_shapes=([None]))

dataset = dataset.repeat(num_epochs)
```

```
 #to see if batches are formed
for raw_record in dataset.take(1):           
  print(repr((raw_record)))
  #print(type(raw_record))
  
   
```

```
hidden_layer = 512
n_time_steps = 200
```

```
#initial state

#initial = tf.Variable(tf.keras.initializers.GlorotNormal([128,512],minval= -0.1,maxval= 0.1,dtype = np.float32))

#input to hidden weight# weight getting multiplied to input
initializer = tf.keras.initializers.GlorotUniform()

w_xh = tf.Variable(initializer(shape = [vocab_size,hidden_layer]))

#hidden to hidden weight # weight getting multiplied to previous hidden out

w_hh = tf.Variable(initializer(shape= [hidden_layer,hidden_layer]))

#bias which is added after input and hidden matmul
b_hh = tf.Variable(tf.random.uniform([hidden_layer],minval= 0,maxval= 0,dtype = np.float32))

#hidden to output weights#weight multiplied after hidden layer
w_h0 = tf.Variable(initializer(shape=[hidden_layer,vocab_size]))

# bias after intermediate and out weights matmul
b_h0 = tf.Variable(tf.random.uniform([vocab_size],minval= 0,maxval= 0,dtype = np.float32))
variable = [w_xh,w_hh,b_hh,w_h0,b_h0]
```

```
opt = tf.optimizers.Adam()
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits= True)
train_acc_metric = tf.keras.metrics.SparseCategoricalAccuracy()
test_acc_metric = tf.keras.metrics.SparseCategoricalAccuracy()
```

```
for dat in dataset:
  
  with tf.GradientTape(persistent=True) as tape:
    losser = tf.TensorArray(size = 200,dtype= tf.float32,dynamic_size= True)
    h_t = tf.Variable(tf.random.uniform([batch_size,hidden_layer],minval= 0,maxval= 0,dtype = np.float32))
    for time_step in tf.range(tf.shape(dat)[1]-1):
      
      data = tf.one_hot(dat[:,time_step], vocab_size)
      #print(tf.shape(data))
      # initial = tf.Variable(tf.random.uniform([batch_size,hidden_layer],minval= 0,maxval= 0,dtype = np.float32))
      # if time_step == 0:
      #   a = (tf.matmul(data,w_xh))+ (tf.matmul(initial,w_hh)) + b_hh
      # else:
      a = (tf.matmul(data,w_xh))+ (tf.matmul(h_t, w_hh)) + b_hh
      h_t = tf.nn.tanh(a)
      logits = (tf.matmul(h_t, w_h0)) + b_h0
      logits = tf.convert_to_tensor(logits)
      # print(tf.shape(logits))
      # print(tf.shape(data[:,time_step+1]))

      loss_val = loss(dat[:,time_step+1],logits)
      #print(loss_val)
      losser.write(time_step,value = loss_val).mark_used()
      
  #print(losser.stack()) 
       
    value = tf.math.reduce_mean(losser.stack())
    print((value))  
  
  grads = tape.gradient(value, [w_xh,w_hh,b_hh,w_h0,b_h0]) 
  #print(grads) 
  opt.apply_gradients(zip(grads,[w_xh,w_hh,b_hh,w_h0,b_h0]))




```
